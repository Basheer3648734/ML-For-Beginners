```python
import pandas as pd
df_students = pd.read_csv('grades.csv',delimiter=',',header='infer')

df_students = df_students.dropna(axis=0)

passes = pd.Series(df_students.Grade >= 60)
df_students = pd.concat([df_students, passes.rename('Pass')], axis=1)
print(df_students)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(8,3))
plt.bar(x= df_students.Name, height=df_students.Grade, color='Orange')
plt.title('Bar chart showing student name and grade')
plt.xlabel('Student name')
plt.ylabel('Grade')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,4))

ax[0].bar(df_students.Name, height=df_students.Grade, color="Orange")
ax[0].set_xticklabels(labels=df_students.Name, rotation=90)
ax[0].set_title('Grades')

pass_count = df_students.Pass.value_counts()
ax[1].pie(pass_count, labels = pass_count)
ax[1].legend(pass_count.keys().tolist())
ax[1].set_title('Passing grades')

fig.suptitle('Student data')

fig.show()
```
